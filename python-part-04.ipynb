Group by - Like a Pivot table or an SQL group by. Let's take this data set and partition it by the chromosome

````python
by_chrom = b.groupby('chrom')
````

````python
by_chrom.size()
````

Can also define a function to group by. Very powerful

````python
b['chromSpan'] = b['chromEnd'] - b['chromStart']
````

Basically our script in one line.

````python
b.head()
````

````python
by_chrom.chromSpan.max()
````

````python
b[b.chrom == 'chrM'][b.chromSpan == 1011]
````

````python
by_chrom.chromSpan.max().plot(kind='bar')
````
